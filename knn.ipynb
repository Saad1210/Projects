## Importing libraries

```python
import pandas as pd
```

## Reading Data

```python
df= pd.read_csv("iitm_intern_train_data.csv")
test=pd.read_csv("iitm_intern_test_data.csv")
```

## getting some more insights of the trainig data using describe method

```python
df.describe()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop("class", axis=1),df['class'],test_size=0.20)
```

```python
X_train
```

### Preprocessing the data

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
scaler.fit(test)
```

```python
scaled_features = scaler.transform(X_train)
scaled_valid = scaler.transform(X_test)
scaled_test = scaler.transform(test)
```

```python
scaled_features
```

```python
df_train = pd.DataFrame(scaled_features,columns=df.columns[:-1])
df_val = pd.DataFrame(scaled_valid,columns=df.columns[:-1])
df_test = pd.DataFrame(scaled_test,columns=df.columns[:-1])
df_train.head()
```

### using gridsearch to found the optimum parameters which is dominating n_neighbours in KNN model

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
```

```python
parameters = {'n_neighbors':[5,10, 20, 50, 80, 100, 500]}

knn = KNeighborsClassifier()
gs = GridSearchCV(knn, parameters, n_jobs=-1)
gs.fit(df_train, y_train)
```

```python
gs.best_params_
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=10)
knn.fit(df_train,y_train)
pred_knn = knn.predict(df_val)
```

```python
from sklearn.metrics import accuracy_score, mean_squared_error
```

```python
mean_squared_error(y_test,pred_knn)
```

### Got the accuracy score of around 82%

```python
print(accuracy_score(y_test,pred_knn))
```
